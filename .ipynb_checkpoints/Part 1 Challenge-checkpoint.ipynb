```python
#import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
```

## Description
The attached logins.json file contains (simulated) timestamps of user logins in a particular geographic location. Aggregate these login counts based on 15minute time intervals, and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand. Please report/illustrate important features of the demand, such as daily cycles. If there are data quality issues, please report them.

```python
# read in and examine logins file
logins = pd.read_json('logins.json')
logins.head()
```

```python
print('The amount of logins given to us for examination are', len(logins.index))
print('First recorded login is', min(logins['login_time']))
print('Last recorded login is', max(logins['login_time']))
print('So we are looking at', max(logins['login_time']) - min(logins['login_time']), ' worth of data' )
```

### The data is in a clean date type format with a little over 3 months worth of data. In order to more easily glean insights from the data we will resample the data at 15 minute intervals.

### Resample at 15 minute counts

```python
#Resample at 15 minute counts
logins15 = pd.DataFrame()
logins['count'] = 1
logins15 = logins.groupby(pd.Grouper(key="login_time", freq="15min")).sum()

logins15.head()
```

```python
plt.figure(figsize=(20,5))
plt.plot(logins15)

plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Amount of logins over a timeperiod')
```

```python
print(' On the graph we can easily see the', max(logins15['count']), ' count login spike in early March.')
```

```python
#The specific timestamp of our high login
logins15.loc[logins15['count']== max(logins15['count'])]
```

The minimums are not as stark and easily summarized since there are several 15 minute time periods with 1 - 0 logins.

### Resample by day

If we summarize by days of the week we can easily find the most and least popular days for logins.

```python
logins1 = logins
logins1['weekday'] = logins['login_time'].dt.dayofweek
```

```python
loginsDaily = logins1.groupby(pd.Grouper(key="weekday")).sum().sort_values('weekday').reset_index()

loginsDaily.head(7)
```

```python
plt.plot(loginsDaily)
plt.xlabel('Day of the week: Mon-Sun')
plt.ylabel('Amount of logins')
plt.title('Amount of logins by day of the week')
```

```python
plt.scatter(loginsDaily['weekday'], loginsDaily['count'])
```
