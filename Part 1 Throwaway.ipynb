My attempt to group by the 15 minute intervals and get information from just the time excluding dates. They were all pretty consistent across the board and didn't yield results.

```python
#import necessary modules
import pandas as pd
```

```python
# read in and examine logins file
logins = pd.read_json('logins.json')
logins.head()
logins.shape
```

```python
grouped15 = pd.DataFrame(logins.groupby([pd.Grouper(key='login_time', freq='15Min')]).sum()).reset_index()

print(grouped15.head())
print(grouped15.tail())
```

```python
timeOnly = pd.DataFrame()
timeOnly['LastDigit'] = grouped15['login_time'].apply(lambda x: str(x)[-8:])
timeOnly.head()
```

```python
frequentTimes = pd.DataFrame()
frequentTimes['occurenceCount'] = timeOnly['LastDigit'].value_counts().sort_values(ascending = False)
```

```python
#Verify the appropriate number of logins
print(frequentTimes['occurenceCount'].sum())
print(frequentTimes.shape)
```

```python
frequentTimes = frequentTimes.reset_index().rename(columns={"index": "time"})
```

```python
frequentTimes = frequentTimes.sort_values(by='occurenceCount', ascending=False)
```
